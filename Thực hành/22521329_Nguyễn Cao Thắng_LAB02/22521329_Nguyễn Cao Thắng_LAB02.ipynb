# Bài 3. Xây dựng lớp PhanSo

```python
from math import gcd

class PhanSo:
    def __init__(self, tu, mau):
        if mau == 0:
            raise ValueError("Mẫu số không thể bằng 0")
        self.tu = tu
        self.mau = mau
        self.original_tu = tu
        self.original_mau = mau
        self.rut_gon()

    def rut_gon(self):
        ucln = gcd(self.tu, self.mau)
        self.tu //= ucln
        self.mau //= ucln
        if self.mau < 0:
            self.tu = -self.tu
            self.mau = -self.mau

    def doi_dau(self):
        self.original_tu = -self.original_tu

    def __add__(self, other):
        if isinstance(other, PhanSo):
            tu = self.tu * other.mau + other.tu * self.mau
            mau = self.mau * other.mau
        else:
            tu = self.tu + other * self.mau
            mau = self.mau
        return PhanSo(tu, mau)

    def __sub__(self, other):
        if isinstance(other, PhanSo):
            tu = self.tu * other.mau - other.tu * self.mau
            mau = self.mau * other.mau
        else:
            tu = self.tu - other * self.mau
            mau = self.mau
        return PhanSo(tu, mau)

    def __mul__(self, other):
        if isinstance(other, PhanSo):
            tu = self.tu * other.tu
            mau = self.mau * other.mau
        else:
            tu = self.tu * other
            mau = self.mau
        return PhanSo(tu, mau)

    def __truediv__(self, other):
        if isinstance(other, PhanSo):
            tu = self.tu * other.mau
            mau = self.mau * other.tu
        else:
            tu = self.tu
            mau = self.mau * other
        return PhanSo(tu, mau)

    def __eq__(self, other):
        return self.tu * other.mau == self.mau * other.tu

    def __ne__(self, other):
        return not self.__eq__(other)

    def __gt__(self, other):
        return self.tu * other.mau > self.mau * other.tu

    def __ge__(self, other):
        return self.tu * other.mau >= self.mau * other.tu

    def __lt__(self, other):
        return self.tu * other.mau < self.mau * other.tu

    def __le__(self, other):
        return self.tu * other.mau <= self.mau * other.tu

    def __str__(self):
        return f'{self.original_tu}/{self.original_mau}'
```

# Bài 4. Xây dựng lớp ToaDo

```python
import math
class Toa_Do:
    def __init__(self, x=0, y=0):
        self.__x = x
        self.__y = y

    def __str__(self):
        return "({0}, {1})".format(self.x, self.y)

    @property
    def x(self):
        return self.__x
    
    @property
    def y(self):
        return self.__y
    
    @x.setter
    def x(self, x):
        self.__x = x
        
    @y.setter
    def y(self, y):
        self.__y = y

    @x.deleter
    def x(self):
        del self.__x

    @y.deleter
    def y(self):
        del self.__y

    def move_to(self, x, y):
        self.x = x
        self.y = y

    def distance_to(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        return math.sqrt(dx*dx + dy*dy)

    def midpoint_to(self, other):
        mx = (self.x + other.x) / 2
        my = (self.y + other.y) / 2
        return Toa_Do(mx, my)

    def delete(self):
        del self.x
        del self.y

    def clone(self):
        return Toa_Do(self.x, self.y)
```

```python
if __name__ == '__main__':
    #1. Khởi tạo tọa độ một điểm
    point1 = Toa_Do(3, 4)
    point2 = Toa_Do(6, 8)

    #2. Xuất tọa độ một điểm
    print(f'Tọa độ điểm 1: {point1}')
    print(f'Tọa độ điểm 2: {point2}')

    #3. Lấy tung độ, hoành độ của một điểm
    print(f'Hoành độ của điểm 1: {point1.x}')
    print(f'Tung độ của điểm 1: {point1.y}')

    #4. Thay đổi tung độ, hoành độ của một điểm
    point1.x = 5
    point1.y = 7
    print(f'Tọa độ điểm 1 sau khi thay đổi: {point1}')

    #5. Di chuyển đến tọa độ
    point1.move_to(10, 10)
    print(f'Tọa độ điểm 1 sau khi di chuyển: {point1}')

    #6. Khoảng cách hai điểm
    distance = point1.distance_to(point2)
    print(f'Khoảng cách giữa điểm 1 và điểm 2: {distance}')

    #7. Tìm tọa độ trung điểm khi biết hai điểm
    midpoint = point1.midpoint_to(point2)
    print(f'Tọa độ trung điểm giữa điểm 1 và điểm 2: {midpoint}')

    #8. Xóa tọa độ một điểm
    point1.delete()
    try:
        print(f'Tọa độ điểm 1 sau khi xóa: {point1}')
    except Exception as e:
        print(f'Đã xảy ra lỗi: {e}')

    #9. Sao chép tọa độ một điểm
    point2_clone = point2.clone()

    #10. Tìm tọa độ có khoảng cách gần tâm O nhất trong nhiều tọa độ
    points = [Toa_Do(3, 4), Toa_Do(1, 1), Toa_Do(0, 2), Toa_Do(2, 2)]
    closest_point = min(points, key=lambda p: p.distance_to(Toa_Do(0, 0)))
    print(f'Tọa độ gần tâm O nhất: {closest_point}')
```

# Bài 5. Thiết kế lớp TamGiac

```python
class Tam_Giac:
    def __init__(self, A, B, C):
        self.__A = A
        self.__B = B
        self.__C = C

    def __str__(self):
        return f'Tam giác với các đỉnh: A{self.__A}, B{self.__B}, C{self.__C}'

    @property
    def A(self):
        return self.__A

    @property
    def B(self):
        return self.__B

    @property
    def C(self):
        return self.__C

    def loai_tam_giac(self):
        AB = self.__A.distance_to(self.__B)
        BC = self.__B.distance_to(self.__C)
        CA = self.__C.distance_to(self.__A)
        if AB == BC == CA:
            return "Tam giác đều"
        elif AB == BC or BC == CA or CA == AB:
            return "Tam giác cân"
        elif AB**2 + BC**2 == CA**2 or BC**2 + CA**2 == AB**2 or CA**2 + AB**2 == BC**2:
            return "Tam giác vuông"
        else:
            return "Tam giác thường"

    def chu_vi(self):
        AB = self.__A.distance_to(self.__B)
        BC = self.__B.distance_to(self.__C)
        CA = self.__C.distance_to(self.__A)
        return AB + BC + CA

    def dien_tich(self):
        AB = self.__A.distance_to(self.__B)
        BC = self.__B.distance_to(self.__C)
        CA = self.__C.distance_to(self.__A)
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - AB) * (p - BC) * (p - CA))

    def ban_kinh_ngoai_tiep(self):
        AB = self.__A.distance_to(self.__B)
        BC = self.__B.distance_to(self.__C)
        CA = self.__C.distance_to(self.__A)
        dien_tich = self.dien_tich()
        return (AB * BC * CA) / (4 * dien_tich)

    def ban_kinh_noi_tiep(self):
        dien_tich = self.dien_tich()
        chu_vi = self.chu_vi()
        return dien_tich / (chu_vi / 2)
```

```python
if __name__ == '__main__':
    #1. Khởi tạo tam giác
    A = Toa_Do(0, 0)
    B = Toa_Do(3, 0)
    C = Toa_Do(0, 4)
    triangle = Tam_Giac(A, B, C)

    #2. Xuất tam giác
    print(triangle)

    #3. Loại tam giác
    print(f'Loại tam giác: {triangle.loai_tam_giac()}')

    #4. Chu vi tam giác
    print(f'Chu vi tam giác: {triangle.chu_vi()}')

    #5. Diện tích tam giác
    print(f'Diện tích tam giác: {triangle.dien_tich()}')

    #6. Bán kính ngoại tiếp
    print(f'Bán kính ngoại tiếp: {triangle.ban_kinh_ngoai_tiep()}')

    #7. Bán kính nội tiếp
    print(f'Bán kính nội tiếp: {triangle.ban_kinh_noi_tiep()}')
```

# Bài 6. Xây dựng một lớp để biểu diễn một đường tròn trong mặt phẳng.

```python
class Duong_Tron:
    def __init__(self, tam, ban_kinh):
        self.__tam = tam
        self.__ban_kinh = ban_kinh

    @property
    def tam(self):
        return self.__tam

    @tam.setter
    def tam(self, tam):
        self.__tam = tam

    @property
    def ban_kinh(self):
        return self.__ban_kinh

    @ban_kinh.setter
    def ban_kinh(self, ban_kinh):
        self.__ban_kinh = ban_kinh

    def __str__(self):
        return f'Đường tròn tâm {self.__tam} và bán kính {self.__ban_kinh}'

    def chu_vi(self):
        return 2 * math.pi * self.__ban_kinh

    def dien_tich(self):
        return math.pi * self.__ban_kinh ** 2

    def phong_to(self, phan_tram):
        self.__ban_kinh *= (1 + phan_tram / 100)

    def thu_nho(self, phan_tram):
        self.__ban_kinh *= (1 - phan_tram / 100)
    
    def giao_nhau(self, other):
        distance_between_centers = self.__tam.distance_to(other.tam)
        if distance_between_centers < (self.__ban_kinh + other.ban_kinh):
            print("Hai đường tròn có giao nhau")
        else:
            print("Hai đường tròn không giao nhau")
```

```python
# Chương trình minh họa
if __name__ == '__main__':
    #1. Khởi tạo đường tròn
    tam = Toa_Do(0, 2)
    duong_tron = Duong_Tron(tam, 5)

    #2. Xuất thông tin đường tròn
    print(duong_tron)

    #3. Tính chu vi
    print(f'Chu vi đường tròn: {duong_tron.chu_vi()}')

    #4. Tính diện tích
    print(f'Diện tích đường tròn: {duong_tron.dien_tich()}')

    #5. Nhập số % phóng to đường tròn
    phan_tram_phong_to = float(input("Nhập số % phóng to đường tròn: "))
    duong_tron.phong_to(phan_tram_phong_to)
    print(f'Đường tròn sau khi phóng to {phan_tram_phong_to}%: {duong_tron}')

    #6. Nhập số % thu nhỏ đường tròn
    phan_tram_thu_nho = float(input("Nhập số % thu nhỏ đường tròn: "))
    duong_tron.thu_nho(phan_tram_thu_nho)
    print(f'Đường tròn sau khi thu nhỏ {phan_tram_thu_nho}%: {duong_tron}')
    
    #7. Kiểm tra hai đường tròn có giao nhau không
    tam1 = Toa_Do(0, 0)
    tam2 = Toa_Do(0, 6)
    circle1 = Duong_Tron(tam1, 5)
    circle2 = Duong_Tron(tam2, 5)
    circle1.giao_nhau(circle2)
```
