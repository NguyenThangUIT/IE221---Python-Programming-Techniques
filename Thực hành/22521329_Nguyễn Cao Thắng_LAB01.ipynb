## Bài 15. Nhập hai số a, b. Sau đó, tính toán biểu thức sau

```python
import math

a = float(input('Nhập số a: '))
b = float(input('Nhập số b: '))

# Tính căn bậc 3 của a, b và ab
cube_root_a = a**(1/3)
cube_root_b = b**(1/3)
cube_root_ab = (a * b)**(1/3)

# Tính biểu thức
numerator = ((a + b)/(cube_root_a + cube_root_b)) - cube_root_ab
denominator = (cube_root_a - cube_root_b)**2
result = numerator / denominator

print(f'Kết quả của biểu thức là: {result}')
```

## Bài 23. Giải và biện luận phương trình bậc hai: ax2 + bx + c = 0

```python
import math

a = float(input('Nhập hệ số a: '))
b = float(input('Nhập hệ số b: '))
c = float(input('Nhập hệ số c: '))
print(f'Phương trình: {a}x^2 + {b}x + {c} = 0')

# Giải và biện luận phương trình bậc hai ax^2 + bx + c = 0
if a == 0:
    if b == 0:
        if c == 0:
            print('Phương trình có vô số nghiệm.')
        else:
            print('Phương trình vô nghiệm.')
    else:
        x = -c / b
        print(f'Phương trình có nghiệm x = {x}')
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print('Phương trình vô nghiệm.')
    elif delta == 0:
        x = -b / (2*a)
        print(f'Phương trình có nghiệm kép x = {x}')
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f'Phương trình có hai nghiệm phân biệt x1 = {x1} và x2 = {x2}')
```

```python
import math

#Chọn hình và nhập kích thước, tính chu vi và diện tích
shape = input('Nhập hình (v: hình vuông, n: hình chữ nhật, t: hình tròn): ')

if shape == 'v':
    side = float(input('Nhập cạnh của hình vuông: '))
    perimeter = 4 * side
    area = side * side
    print(f'Chu vi hình vuông: {perimeter}')
    print(f'Diện tích hình vuông: {area}')
elif shape == 'n':
    width = float(input('Nhập chiều rộng: '))
    length = float(input('Nhập chiều dài: '))
    perimeter = 2 * (width + length)
    area = width * length
    print(f'Chu vi hình chữ nhật: {perimeter}')
    print(f'Diện tích hình chữ nhật: {area}')
elif shape == 't':
    radius = float(input('Nhập bán kính của hình tròn: '))
    perimeter = 2 * math.pi * radius
    area = math.pi * radius * radius
    print(f'Chu vi hình tròn: {perimeter}')
    print(f'Diện tích hình tròn: {area}')
else:
    print('Hình không hợp lệ')
```

## Bài 54. Anh/chị nhập vào một số nguyên dương n và viết các phương thức sau:

```python
import math

def calc_root(n, x):
    """
    Tính căn bậc x của số n.
    """
    return n ** (1 / x)

def reverse_number(n):
    """
    Trả về số đảo của n.
    """
    return int(str(n)[::-1])

def is_perfect_square(n):
    """
    Kiểm tra xem n có phải là số chính phương hay không.
    """
    return int(math.sqrt(n)) ** 2 == n

def is_prime(n):
    """
    Kiểm tra xem n có phải là số nguyên tố hay không.
    """
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def product_of_odd_digits(n):
    """
    Tính tích các chữ số lẻ của n.
    """
    product = 1
    has_odd = False
    for digit in str(n):
        if int(digit) % 2 != 0:
            product *= int(digit)
            has_odd = True
    return product if has_odd else 0

def sum_of_primes_less_than(n):
    """
    Tính tổng các số nguyên tố nhỏ hơn n.
    """
    return sum(i for i in range(2, n) if is_prime(i))

def sum_of_perfect_squares_less_than(n):
    """
    Tính tổng các số chính phương nhỏ hơn n.
    """
    return sum(i for i in range(1, n) if is_perfect_square(i))

def sum_of_divisors(n):
    """
    Tính tổng các ước số dương của n.
    """
    return sum(i for i in range(1, n + 1) if n % i == 0)

# Nhập n và x từ bàn phím
n = int(input('Nhập số nguyên dương n: '))
x = int(input('Nhập số nguyên dương x: '))
print(f'Căn bậc {x} của {n} là: {calc_root(n, x)}')
print(f'Số đảo của {n} là: {reverse_number(n)}')
print(f'{n} có phải là số chính phương không? {is_perfect_square(n)}')
print(f'{n} có phải là số nguyên tố không? {is_prime(n)}')
print(f'Tích các chữ số lẻ của {n} là: {product_of_odd_digits(n)}')
print(f'Tổng các số nguyên tố nhỏ hơn {n} là: {sum_of_primes_less_than(n)}')
print(f'Tổng các số chính phương nhỏ hơn {n} là: {sum_of_perfect_squares_less_than(n)}')
print(f'Tổng các ước số dương của {n} là: {sum_of_divisors(n)}')
```

## Bài 55. Nhập vào một số nguyên dương n và viết các phương thức sau

```python
import math

def sum_of_numbers(n):
    """
    n (int): Số nguyên dương.
    Trả về tổng các số từ 1 đến n.
    """
    return sum(range(1, n + 1))

def sum_with_2_appended(n):
    """
    n (int): Số nguyên dương.
    Trả về tổng các số với chữ số 2 chèn thêm sau mỗi số.
    """
    total = 0
    for i in range(1, n + 1):
        total += int(f"{i}2")
    return total

def harmonic_sum(n):
    """
    n (int): Số nguyên dương.
    Trả về tổng các phân số từ 1 đến 1/n.
    """
    return sum(1 / i for i in range(1, n + 1))

def factorial_sum(n):
    """
    n (int): Số nguyên dương.
    Trả về tổng các giai thừa từ 1! đến n!.
    """
    return sum(math.factorial(i) for i in range(1, n + 1))

def product_of_numbers(n):
    """
    n (int): Số nguyên dương.
    Trả về tích các số từ 1 đến n.
    """
    product = 1
    for i in range(1, n + 1):
        product *= i
    return product
    
n = int(input("Nhập một số nguyên dương n: "))
print(f"Tổng các số từ 1 đến {n}: {sum_of_numbers(n)}")
print(f"Tổng các số mà chèn thêm chữ số 2 vào sau mỗi số từ 1 đến {n}: {sum_with_2_appended(n)}")
print(f"Tổng các phân số từ 1 đến 1/{n}: {harmonic_sum(n)}")
print(f"Tổng các giai thừa từ 1! đến {n}!: {factorial_sum(n)}")
print(f"Tích các số từ 1 đến {n}: {product_of_numbers(n)}")
```

## Bài 56. Viết phương thức in ra n phần tử của dãy Fibonacci

```python
def fibonacci_sequence(n):
    """
    Trả về danh sách chứa n phần tử đầu tiên của dãy Fibonacci.
    """
    fib_sequence = [0, 1]
    while len(fib_sequence) < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]

# In ra n phần tử của dãy Fibonacci
n = int(input('Nhập số phần tử cần in của dãy Fibonacci: '))
fib_sequence = fibonacci_sequence(n)
print(f'{n} phần tử đầu tiên của dãy Fibonacci là: {fib_sequence}')
```

## Bài 57. Viết phương thức tính chu vi và diện tích hình chữ nhật khi biết độ dài 2 cạnh. Sau đó vẽ hình chữ nhật ra màn hình bằng các dấu *. Phương thức tính chu vi, diện tích và phương thức vẽ hình chữ nhật phải độc lập nhau.

```python
def calculate_perimeter(a: int, b: int) -> int:
    return 2 * (a + b)

def calculate_area(a: int, b: int) -> int:
    return a * b

def draw_rectangle(a: int, b: int):
    for i in range(b):
        if i == 0 or i == b - 1:
            print('* ' * a)
        else:
            print('*' + ' ' * (2 * (a - 2) + 1) + '*')

# Nhập hai tham số a và b từ người dùng
a = int(input('Nhập chiều dài a: '))
b = int(input('Nhập chiều rộng b: '))

print(f'Chu vi: {calculate_perimeter(a, b)}')
print(f'Diện tích: {calculate_area(a, b)}')
print('Hình chữ nhật:')
draw_rectangle(a, b)
```

## Bài 58. Áp dụng kỹ thuật kwargs, *args, ** kwargs thiết kế hàm tính chu vi, diện tích hình vuông, hình chữ nhật, hình tròn theo các phương pháp gọi hàm sau:

```python
import math

# Hàm tính chu vi, diện tích dựa theo tham số hình và tác vụ cần thực hiện:
def tinh(*args, **kwargs):
    hinh = kwargs.get('hinh')
    tinh = kwargs.get('tinh')
    
    if hinh == 'vuong':
        canh = args[0]
        if tinh == 'cv':
            return f'Chu vi của hình vuông là: {4 * canh}'
        elif tinh == 'dt':
            return f'Diện tích của hình vuông là: {canh * canh}'
    
    elif hinh == 'chu_nhat':
        dai, rong = args
        if tinh == 'cv':
            return f'Chu vi của hình chữ nhật là: {2 * (dai + rong)}'
        elif tinh == 'dt':
            return f'Diện tích của hình chữ nhật là: {dai * rong}'
    
    elif hinh == 'tron':
        ban_kinh = args[0]
        if tinh == 'cv':
            return f'Chu vi của hình tròn là: {2 * math.pi * ban_kinh}'
        elif tinh == 'dt':
            return f'Diện tích của hình tròn là: {math.pi * ban_kinh * ban_kinh}'

    else:
        return 'Input không hợp lệ'
```

```python
print(tinh(10, hinh='vuong', tinh='cv'))
print(tinh(50, hinh='vuong', tinh='dt'))
print(tinh(18, hinh='tron', tinh='cv'))
print(tinh(20, 30, hinh='chu_nhat', tinh='cv'))
```

```python
#Xuất thông tin sinh viên có điểm trung binh cao nhất:
max_gpa = max([student['gpa'] for student in students])
for student in students:
    if student['gpa'] == max_gpa:
        print(f'Sinh viên có điểm trung bình cao nhất: {student['name']} với điểm GPA là: {student['gpa']}')
```

```python
# Cập nhật xếp loại học lực cho từng sinh viên trong danh sách
def update_classification(students):
    for student in students:
        gpa = student['gpa']
        if gpa < 3.5:
            student['classification'] = 'Kém'
        elif 3.5 <= gpa < 5.0:
            student['classification'] = 'Yếu'
        elif 5.0 <= gpa < 7.0:
            student['classification'] = 'Trung bình'
        elif 7.0 <= gpa < 8.0:
            student['classification'] = 'Khá'
        elif 8.0 <= gpa < 9.0:
            student['classification'] = 'Giỏi'
        elif 9.0 <= gpa <= 10:
            student['classification'] = 'Xuất sắc'
    return students
# Gọi phương thức để cập nhật xếp loại học lực
update_classification(students)
```

```python
#Tìm sinh viên theo mã sinh viên:
def find_student_by_id(student_id):
    for student in students:
        if student['student_id'] == student_id:
            print(f'Thông tin sinh viên với mã {student_id}: \nMã sinh viên: {student['student_id']}, Tên: {student['name']}, GPA: {student['gpa']}, Xếp loại: {student['classification']}')
            return student
    print(f'Không tìm thấy sinh viên với mã {student_id}')
    return None

find_studen_1 = find_student_by_id('SV007')
```

```python
# Tìm tất cả sinh viên theo điểm trung bình:
def find_students_by_gpa(gpa):
    found_students = []
    for student in students:
        if student['gpa'] == gpa:
            found_students.append(student)
    if found_students:
        print(f'Thông tin sinh viên có điểm GPA là {gpa}:')
        for student in found_students:
            print(f'Mã sinh viên: {student["student_id"]}, Tên: {student["name"]}, GPA: {student["gpa"]}, Xếp loại: {student["classification"]}')
    else:
        print(f'Không tìm thấy sinh viên có điểm GPA là {gpa}')
    return found_students

find_students_2 = find_students_by_gpa(9.75)
```
